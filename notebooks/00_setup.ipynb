## 6. Matplotlib ÊµãËØï ‚Äî ÁªòÂõæ

```python
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Test Plot: y = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
